```python
# Bogo_Sort

import random

# Sorts array a[0..n-1] using Bogo sort
def bogoSort(a):
    n = len(a)
    while not is_sorted(a):
        shuffle(a)
        print("Current array state:", a)

# To check if array is sorted or not
def is_sorted(a):
    n = len(a)
    for i in range(0, n-1):
        if a[i] > a[i+1]:
            return False
    return True

# To generate permutation of the array
def shuffle(a):
    n = len(a)
    for i in range(n):
        r = random.randint(0, n-1)
        a[i], a[r] = a[r], a[i]

# Driver code to test above
a = [3, 2, 4, 0]
bogoSort(a)
print("=======================")
print("Sorted array:")
print(" ".join(str(x) for x in a))
print("=======================")
##########################################################

# Bubble_sort

def bubbleSort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n-i-1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
            # Print the array after each swap
            print("Step:", arr)
        print("After pass", i+1, ":", arr)
        print("")
    return arr


# Driver code to test above
arr = [64, 34, 25, 12, 22, 11, 90]
print("Original array:", arr)
sorted_arr = bubbleSort(arr)
print("Sorted array:", sorted_arr)
##########################################################

# Selection_Sort

def selectionSort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Find the minimum element in remaining unsorted array
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j

        # Swap the found minimum element with the first element of the unsorted part
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

        # Print the array after each iteration of selection sort
        print("After iteration", i+1, ":", arr)

    return arr


# Driver code to test above
arr = [64, 25, 12, 22, 11]
print("Original array:", arr)
sorted_arr = selectionSort(arr)
print("Sorted array:", sorted_arr)
##########################################################

# Part B


a = [8, 3, 7, 4, 2]
print("Original array:", a)
bogoSort(a)
print("=======================")
print("Sorted array:")
print(" ".join(str(x) for x in a))
print("=======================")
##########################################################

arr = [8, 3, 7, 4, 2]
print("Original array:", arr)
sorted_arr = bubbleSort(arr)
print("Sorted array:", sorted_arr)
##########################################################

arr = [8, 3, 7, 4, 2]
print("Original array:", arr)
sorted_arr = selectionSort(arr)
print("Sorted array:", sorted_arr)
##########################################################

# Part C

# 1) Buuble Sort and Selection Sort are much faster then the Bogo Sort.
# 2) Bogo Sort is the slowest. It relies on random shuffling until the array happens to be sorted, which is inefficent.
# 3) Bogo Sort's significant lack of time complexity and dependence on pure chance makes it unsuitable when compared to Bubble and Selection Sort.
```
